```python
# Install required packages if you haven't already
# !pip install fuzzywuzzy python-Levenshtein pandas

from fuzzywuzzy import fuzz
from fuzzywuzzy import process

# Method 1: Create a mapping dictionary using fuzzy matching
def get_best_match(name, list_names, min_score=0):
    # Get the best match for name from list_names
    match = process.extractOne(name, list_names)
    if match[1] >= min_score:
        return match[0]
    else:
        return None

# Create a mapping between df2 names and df1 names
name_mapping = {}
for name in PLUTODF['company_name']:
    match = get_best_match(name, df1['name'].tolist(), min_score=70)
    if match:
        name_mapping[name] = match

# Add a column to df2 with the matched name from df1
df2['matched_name'] = df2['company_name'].map(name_mapping)

# Now merge using the matched name
result = pd.merge(df2, df1, left_on='matched_name', right_on='name', how='left')

# Method 2: For more complex scenarios, create a custom function
def fuzzy_merge(df_left, df_right, left_on, right_on, threshold=90, limit=1):
    """
    Performs a fuzzy merge of two dataframes based on specified columns
    
    Parameters:
    -----------
    df_left : pandas DataFrame
        Left dataframe
    df_right : pandas DataFrame
        Right dataframe
    left_on : str
        Column name from left dataframe to use for matching
    right_on : str
        Column name from right dataframe to use for matching
    threshold : int, optional (default=90)
        Minimum similarity score (0-100) to consider a match
    limit : int, optional (default=1)
        Maximum number of matches to return per row
        
    Returns:
    --------
    pandas DataFrame
        Merged dataframe
    """
    # Create a copy of right df to avoid modifying the original
    df_right_copy = df_right.copy()
    
    # Create a new column with the match score and matched value
    matches = []
    
    # For each value in the left dataframe
    for left_value in df_left[left_on]:
        # Find potential matches in the right dataframe
        potential_matches = process.extract(
            left_value, 
            df_right_copy[right_on], 
            limit=limit,
            scorer=fuzz.token_sort_ratio
        )
        
        # Filter matches above threshold
        good_matches = [match for match in potential_matches if match[1] >= threshold]
        
        if good_matches:
            best_match = good_matches[0]
            matches.append((left_value, best_match[0], best_match[1]))
        else:
            matches.append((left_value, None, None))
    
    # Create a mapping dataframe
    mapping = pd.DataFrame(matches, columns=[left_on, f'matched_{right_on}', 'score'])
    
    # Merge with the original left dataframe
    df_left_with_matches = pd.merge(df_left, mapping, on=left_on, how='left')
    
    # Merge with the right dataframe
    result = pd.merge(
        df_left_with_matches, 
        df_right_copy, 
        left_on=f'matched_{right_on}', 
        right_on=right_on, 
        how='left'
    )
    
    return result

# Example usage of the fuzzy_merge function
result = fuzzy_merge(df2, df1, 'company_name', 'name', threshold=70)
```
