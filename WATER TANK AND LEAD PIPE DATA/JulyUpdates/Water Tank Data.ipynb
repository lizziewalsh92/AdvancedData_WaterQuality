AUDIT ANALYSIS

```python
## audits are only updated from 2021 to 3/11/24
df_audits.info()
```

```python
##normalize addresses like in water tank data
df_audits['ADDRESS'] = df_audits['House #'].astype(str)+ ' ' + df_audits['Street Name'].astype(str)
```

```python
df_audits['ADDRESS_NORM'] = df_audits['ADDRESS'].str.lower().str.strip()
```

```python
df_audits_sortedAddress = df_audits.sort_values('Date_of_Occurrence', ascending=False)
```

```python
##sort by BIN number in order to drop all less-recent records
df_audits_sortedBIN = df_audits_sortedAddress.sort_values('BIN', ascending=False)
```

```python
df_audits_DroppedDupeBIN = df_audits_sortedBIN.drop_duplicates(subset='BIN')
```

```python
df_audits_DroppedDupeBIN.sample(5)
```

```python
## remove xtra columns not needed
df_audits_DroppedDupeBIN_NoExtraCols = df_audits_DroppedDupeBIN.drop(['House #', 'Street Name', 'ADDRESS'], axis=1)
```

```python
df_audits_DroppedDupeBIN_NoExtraCols = df_audits_DroppedDupeBIN_NoExtraCols.reset_index(drop=True)
```

```python
## there have been 5402 audit logs between 2021 to 3/11/24
df_audits_DroppedDupeBIN_NoExtraCols.info()
```

```python
df_merged = pd.merge(df_audits_DroppedDupeBIN_NoExtraCols, DF_NODUPES_RENAMEDCOLUMNS, on='BIN', how='outer', indicator=True)
```

```python
df_merged.sample(2)
```

```python
df_merged.to_csv("07.08_merged water tank audits and reports.csv")
```
